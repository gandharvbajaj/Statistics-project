#Outlier detection and handeling



```python
from scipy import stats
important_columns = ['Income', 'MntFishProducts', 'MntMeatProducts', 'MntFruits', 'MntSweetProducts', 'MntWines', 'MntGoldProds', 'NumDealsPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebPurchases', 'NumWebVisitsMonth', 'Recency']


# Calculating the z-score for each value in the 'important_columns' columns
z_scores = stats.zscore(new[important_columns])

# Define a threshold for outlier identification
threshold = 3

# Identify outliers based on the z-score threshold
outliers = new[important_columns][abs(z_scores) > threshold]

# Print the outliers
print(outliers)
```

```python
#  creating boxplot for outlier in important column

for column in important_columns:
  sns.boxplot(x=new[column])
  plt.show()
```
